```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Salary.csv')
x = df['YearsExperience'].values
y = df['Salary'].values
num = len(x)

print(f"there are {num} number of sets of pairs to go through")
cost_history_set = []

class LinearRegression:
    def __init__(self, iterations = 20, alpha = 0.01):
        self.iterations = iterations
        self.theta0 = 0
        self.theta1 = 0
        self.alpha = alpha
    
    def hypothesis(self, x):
        return self.theta0 + self.theta1 * x
    
    def cost(self, x, y):
        predictions = self.hypothesis(x) #gets the y predicted value
        error = predictions - y
        costt = (1/(2*num)) * np.sum(error ** 2)
        return costt
    
    def fit(self, x, y):
        for i in range(self.iterations):
            predictions = self.hypothesis(x)
            errors = predictions - y
            
            grad_theta0 = (1/num) * np.sum(errors)
            grad_theta1 = (1/num) * np.sum(errors * x)
            
            self.theta0 -= grad_theta0 * self.alpha
            self.theta1 -= grad_theta1 * self.alpha
            cost_history_set.append(self.cost(x, y))
            
model = LinearRegression()
model.fit(x, y)


plt.figure(figsize=(10, 6))
plt.plot(cost_history_set)
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE Equation)")
plt.title("Hopefully bowl shaped cost function lmaoo")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize = (10, 6))
plt.scatter(x, y, color = 'red', alpha = 0.6, s=50, label = 'kaggle data')
linex = np.linspace(min(x), max(x), 100)
liney = model.hypothesis(linex)
plt.plot(linex, liney, label = 'l')



            
            
            
            
            
        
    
        
        

print(x)
print(df)
```
